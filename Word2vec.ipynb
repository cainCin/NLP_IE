```python
from wikipedia2vec import Wikipedia2Vec
```

```python
wiki2vec = Wikipedia2Vec.load('jawiki_20180420_100d.pkl')
```

```python
s = '日本語（にほんご、にっぽんご[注 1]）は、主に日本国内や日本人同士の間で使用されている言語である。'
```

```python
print(wiki2vec.get_word_vector(s[0]).shape)
print(len(s))
```

```python
print(wiki2vec.get_word_vector(s).shape)
```

```python
for w in s:
    vector = wiki2vec.get_word_vector(w)
    print(vector)
```
